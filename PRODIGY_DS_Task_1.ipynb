```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Student_performance_data _.csv',  encoding='utf-8')
df.head()
```

```python
data = pd.DataFrame(df)

# Prepare data for visualization
gender_counts = data['Gender'].value_counts()

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index.map({0: 'Male', 1: 'Female'}), gender_counts.values, color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.title('Gender Distribution of High School Students')
plt.show()
```

```python
data = pd.DataFrame(df)

# Prepare data for visualization
gpa_data = data['GPA']

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(gpa_data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('GPA')
plt.ylabel('Number of Students')
plt.title('Distribution of GPA among High School Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
data = pd.DataFrame(df)

# Prepare data for visualization
study_time_data = data['StudyTimeWeekly']

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(study_time_data, bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Study Time Weekly (hours)')
plt.ylabel('Number of Students')
plt.title('Distribution of Weekly Study Time among High School Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
